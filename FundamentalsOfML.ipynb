# IMPORT LIBRARIES

```python
import pandas as pd
data = pd.read_csv('spam_detection_example.csv',encoding='latin-1')
data
print(type(data))
```

# HANDS ON DATAFRAME

```python
print(data.head())
print(type(data))
column = data.columns
print(column)
y = data['label']
print(y)
x = data['message']
print(x)
```

```python
print(data.info())
data['label'] = data['label'].map({'ham':0,'spam':1})
data
```

# SPLIT DATAFRAME ONTO TRAINING AND TESTING SETS

```python
#Split df onto training and testing x & y
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
print(x_train)
print(y_train)
```

# (NPL) NATURAL PROCESSING LANGUAGE - Vectorizing words

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
x_train_count = vectorizer.fit_transform(x_train.values)
print(x_train_count.toarray())
print(x_train_count)
#Now test set
x_test_count = vectorizer.transform(x_test.values)
print(x_test_count.toarray())
print(x_test_count)
```

# TRAINING THE MODEL

```python

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count,y_train)
y_pred = model.predict(x_test_count)
print(y_pred)
print(y_test)
```

# OTHER METHODS




```python
# Import necessary libraries
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KDTree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# Convert sparse matrix to array for KDTree (optional, depending on your approach)
X_train_array = x_train_count.toarray()

# Train a KDTree for nearest neighbor search
kdt = KDTree(X_train_array, leaf_size=30, metric='euclidean')

# Query the KDTree to find the nearest neighbors for the test set
# We are using k=3 as an example
distances, indices = kdt.query(x_test_count.toarray(), k=3)

# For each test sample, find the most frequent label (majority vote)
y_pred = []
for idx in indices:
    labels = y_train.iloc[idx]  # Extract labels using .iloc (index-based)
    majority_label = labels.mode()[0]  # Get the most frequent label (mode)
    y_pred.append(majority_label)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Lista de modelos a evaluar
models = {
    "Naive Bayes": MultinomialNB(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Classifier": SVC()
}

# Entrenamiento y evaluación de cada modelo
for model_name, model in models.items():
    print(f"\nModelo: {model_name}")
    model.fit(x_train_count, y_train)
    y_pred = model.predict(x_test_count)

    # Mostrar resultados
    print("X_test:", x_test)
    print("y_test:", y_test.tolist())
    print("y_pred:", y_pred.tolist())
```

# ACURACCY METRICS

```python
# Diccionario para almacenar métricas
metrics = {}

# Iterar sobre cada modelo
for model_name, model in models.items():
    # Predecir con el modelo
    y_pred = model.predict(x_test_count)

    # Calcular métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, pos_label='spam', zero_division=0)
    recall = recall_score(y_test, y_pred, pos_label='spam', zero_division=0)
    f1 = f1_score(y_test, y_pred, pos_label='spam', zero_division=0)

    # Guardar métricas en el diccionario
    metrics[model_name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1
    }

# Mostrar las métricas de cada modelo
for model_name, metric_values in metrics.items():
    print(f"\nModelo: {model_name}")
    for metric, value in metric_values.items():
        print(f"{metric}: {value:.4f}")
```
